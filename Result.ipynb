## Codebook

`megafon.csv` содержит следующие значения: <br><br>
&nbsp;&nbsp;&nbsp;&nbsp; `user_id` — идентификатор абонента;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Q1` — ответ на первый вопрос;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Q2` — ответ на второй вопрос;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Total Traffic(MB)` — объем трафика передачи данных <sup>1 </sup>; <br>
&nbsp;&nbsp;&nbsp;&nbsp; `Downlink Throughput(Kbps)` — средняя скорость «к абоненту» <sup>2 </sup>;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Uplink Throughput(Kbps)`— средняя скорость «от абонента» <sup>3 </sup>;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Downlink TCP Retransmission Rate(%)` — частота переотправок пакетов «к абоненту» <sup>4 </sup>;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Video Streaming Download Throughput(Kbps)` — скорость загрузки потокового видео <sup>5 </sup>;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Video Streaming xKB Start Delay(ms)` — задержка старта воспроизведения видео <sup>6 </sup>;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Web Page Download Throughput(Kbps)` — скорость загрузки web-страниц через браузер <sup>7 </sup>;<br>
&nbsp;&nbsp;&nbsp;&nbsp; `Web Average TCP RTT(ms)` — пинг при просмотре web-страниц<sup>8 </sup>.<br>


<sup>1 </sup> — Насколько активно абонент использует мобильный интернет.<br>
<sup>2 </sup> — Считается по всему трафику передачи данных.<br>
<sup>3 </sup> — Считается по всему трафику передачи данных.<br>
<sup>4 </sup> — Чем выше, тем хуже. Если в канале возникает ошибка, пакет переотправляется. Снижается полезная скорость.<br>
<sup>5 </sup> — Чем выше, тем лучше — меньше прерываний и лучше качество картинки.<br>
<sup>6 </sup> — Сколько времени пройдёт между нажатием на кнопку Play и началом воспроизведения видео. Чем меньше это время, тем быстрее начинается воспроизведение.<br>
<sup>7 </sup> — Чем выше, тем лучше.<br>
<sup>8 </sup> — Чем меньше, тем лучше — быстрее загружаются web-страницы.<br>

Первый технический показатель представлен как сумма за период в одну неделю перед участием в опросе. Остальные технические показатели отображают среднее значение по данному признаку за период в одну неделю перед участием в опросе.

## 2.1

**Первый и самый очевидный вопрос, ответ на который хотел бы увидеть бизнес по результатам подобного опроса - что является основным техническим драйвером низкой удовлетворенности клиентов качеством услуг. Иными словами, есть ли статистически значимое различие объективных параметров передачи данных у клиентов, поставивших высокие и низкие оценки. Или дело в чем-то другом?**  

```python
#Для ответа на этот вопрос аггрегируем данные по поставленной оценке и смотрим на значения параметров у каждой группы:
df1 = df_no_outliers.groupby('Q1').agg("median")[df_no_outliers.iloc[:,1:9].columns].reset_index()
df1
```

```python
#Проверим корреляцию каждого параметра, чтобы оставить только существующие связи
df1.iloc[:,1:9].corrwith(df1.Q1).sort_values()
```

Видим, что связь присутвует по всем параметрам кроме объема потребенного трафика, чем выше скорость загрузки тем
в среднем выше удовлетворенность потребителей, чем выше задержки в сети, тем удовлетворенность ниже.
Объем потребленного трафика же выше у респондентов, поставивших средние оценки. Удивляют выбивающиеся из общей картины характеристики у пользователей, поставивших 6 баллов, однако, не меняющие общей картины

```python
fig, axes = plt.subplots(figsize = (5,30), nrows = 8)

columns = df1.iloc[:,1:].columns
columns
for i, _ in enumerate(columns):
    axes[i].plot(df1.Q1, df1[columns[i]])
    axes[i].set_title(columns[i])

fig.suptitle('Зависимость оценки от медианного значения параметра', y=0.9, size=16)
plt.subplots_adjust(hspace=0.3)
plt.show()
```

**Сформулируем ряд нулевых гипотез (по одной для каждого параметра), которые проверим с помощью статистических тестов**
В качестве двух популяций рассмотрим записи со значениями оценки (1-5) и (6-10)

Нулевая гипотеза: Технические характеристики связи (каждого параметра по отдельности) не влияют на удовлетворенность клиентов
Контрольным способом будем считать записи со значениями оценки (1-5), а тестовым (6-10).

В качестве теста выберем Two Sample Z-Test. Уровень значимости установим в 1%

```python
for i, _ in enumerate(columns):
    OptionA = df_no_outliers[df_no_outliers['Q1']<=5][columns[i]]
    OptionB = df_no_outliers[df_no_outliers['Q1']>5][columns[i]]
    res = ztest(OptionA, OptionB, value=0)
    if res[1] < 0.01:
        statement="yes"
    else:
        statement="no"
    print(columns[i])
    print("p-value :",res[1])
    print("-----")
```

**Как и преполагалось на основе полученных ранее данных о средних значениях и корреляциях все представленные параметры связи (кроме объема передачи данных) с 99% веротностью имеют крайне значимую связь с итоговой оценкой клиентом услуг оператора. Улучшение качества данных параметров приведет к повышению общей удовлетворенности клиентов**

T-тест – это тест, который сравнивает два средних значения, и сообщает нам, отличаются они друг от друга или нет. Он также показывает, насколько значительны различия. Другими словами, это дает нам представление о том, могли ли эти различия возникнуть случайно.
С помощью опции equal_var = False он выполняет t-тест Уэлча, который не предполагает равную дисперсию совокупности.

Нулевая гипотеза H0: средние значения равны  
Альтернативная гипотеза H1: средние не равны

## 2.3

**Отвечая на второй вопрос, много клиентов отметило проблему низкой скорости. Однако, это субъективное мнение, которое необходимо подтвердить на реальных данных. Предлагается проверить, есть ли статистически значимое различие в ответах пользователей с разными скоростями (Uplink/Downlink Throughput)**

```python
df_uplink = df_no_outliers.loc[:,['Uplink Throughput(Kbps)','slow_internet']]
df_downlink = df_no_outliers.loc[:,['Downlink Throughput(Kbps)','slow_internet']]
```

Для более точного понимания разделим весь массив скоростей на интервалы и посчитаем пропорцию отметивших низкую скорость для данного интервала, затем, сделав вывод о значимости данного критерия для оценки скоростей пользователями

```python
#Число интервалов приближенно определяется по формуле Стерджесса:
m = 1 + 3.322 * np.log10(df_uplink.slow_internet.size)
intervals = int(round(m,0))
```

```python
df_uplink['interval_val'] = pd.qcut(df_uplink['Uplink Throughput(Kbps)'], q=intervals)
df_uplink['interval_id'] = pd.qcut(df_uplink['Uplink Throughput(Kbps)'], q=intervals, labels=range(1,intervals+1))
df_uplink[['interval_val','interval_id']].value_counts().reset_index().sort_values('interval_id')
#получились интервалы примерно по 500 значений
```

```python
#Посчитаем долю пожаловавшихся из каждого интервала
df_uplink2 = df_uplink.groupby(['interval_id']).agg(({'slow_internet':'sum','interval_val':'count'}))
df_uplink2['ratio'] = df_uplink2['slow_internet'] / df_uplink2['interval_val']
df_uplink2
```

```python
#Повторим тоже самое для downlink
df_downlink['interval_val'] = pd.qcut(df_downlink['Downlink Throughput(Kbps)'], q=intervals)
df_downlink['interval_id'] = pd.qcut(df_downlink['Downlink Throughput(Kbps)'], q=intervals, labels=range(1,intervals+1))
df_downlink[['interval_val','interval_id']].value_counts().reset_index().sort_values('interval_id')
```

```python
df_downlink2 = df_downlink.groupby(['interval_id']).agg(({'slow_internet':'sum','interval_val':'count'}))
df_downlink2['ratio'] = df_downlink2['slow_internet'] / df_downlink2['interval_val']
df_downlink2
```

```python

fig, ax = plt.subplots(figsize = (12,6))
ax.plot(df_downlink2.index, df_downlink2['ratio'], label = 'Downlink')
ax.plot(df_uplink2.index, df_uplink2['ratio'], label = 'Uplink')
ax.set_title("Доля пользователей, отметивших низкую скорость для каждой категории скоростей")
ax.set_xlabel('Категория')
ax.set_ylabel('Доля')
ax.set_xticks(range(1,intervals+1))
ax.legend()
plt.show()

```

Из графиков видно, что наиболее сильно процент недовольных скоростью падает по мере роста средней скорости, что выглядит на первый взгляд логично, проверим это утверждение:

Нулевая гипотеза H0: при переходе из категории в категорию процент недовольных скоростью не изменяется
Альтернативная гипотеза H1: процент недовольных скоростью изменяется т категории к категории

```python

for i in range(1,intervals):
    control = df_uplink[df_uplink.interval_id==i]
    test = df_uplink[df_uplink.interval_id==i+1]
    ns_control = control['slow_internet'].sum()
    ns_test = test['slow_internet'].sum()
    n_control = control['slow_internet'].size
    n_test = test['slow_internet'].size
    count = np.array([ns_control, ns_test])
    nobs = np.array([n_control, n_test])
    pval = proportions_ztest(count, nobs)[1]
    if pval < 0.05:
        res = 'опровергнута'
    else:
        res = 'принята'
    print(f'{i} vs {i+1}:')
    print(f'P-value составил {pval:.3f}, а значит нулевая гипотеза {res} на 95% уровне значимости')
    (lower_con, lower_test), (upper_con, upper_test) = proportion_confint(count, nobs=nobs, alpha=0.05)
    print(f'ci 95% для контрольной группы: [{lower_con:.3f}, {upper_con:.3f}]')
    print(f'ci 95% для тестовой группы: [{lower_test:.3f}, {upper_test:.3f}]')  
    print('------')
```

```python
for i in range(1,intervals):
    control = df_downlink[df_downlink.interval_id==i]
    test = df_downlink[df_downlink.interval_id==i+1]
    ns_control = control['slow_internet'].sum()
    ns_test = test['slow_internet'].sum()
    n_control = control['slow_internet'].size
    n_test = test['slow_internet'].size
    count = np.array([ns_control, ns_test])
    nobs = np.array([n_control, n_test])
    pval = proportions_ztest(count, nobs)[1]
    if pval < 0.05:
        res = 'опровергнута'
    else:
        res = 'принята'
    print(f'{i} vs {i+1}:')
    print(f'P-value составил {pval:.3f}, а значит нулевая гипотеза {res} на 95% уровне значимости')
    (lower_con, lower_test), (upper_con, upper_test) = proportion_confint(count, nobs=nobs, alpha=0.05)
    print(f'ci 95% для контрольной группы: [{lower_con:.3f}, {upper_con:.3f}]')
    print(f'ci 95% для тестовой группы: [{lower_test:.3f}, {upper_test:.3f}]')  
    print('------')
```

**По результатам тестов оказалось, что нулевая теория была подтверждена для всех категорий скоростей для Uplink скорости, тогда как для Downlink статистически подтвержденное различие при уровне значимости 95% наблюдается между категориями 1 и 2 (также между 6 и 7, но это является следствием аномального всплеска жалоб для клиентов из категории 6).
